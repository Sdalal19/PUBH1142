<span style="color: violet;">1. What is the difference between a spreadsheet and a database?</span>
 * Spreadsheet:
    * Flat data structure (tabular format).
    * Suitable for small datasets.
    * Limited scalability and integrity features.
 * Database:
   * Complex data structure with tables and relationships.
   * Designed to handle large datasets.
   * Supports advanced querying, multi-user access, and strong data integrity.



<span style="color: violet;">2. What are the advantages and disadvantages of using a spreadsheet for health data management? </span>
* Advantages:
    * Easy to use with minimal training.
    * Allows for quick data entry, sorting, filtering, and visualization.
    * Ideal for small datasets and simple analysis tasks.

* Disadvantages: 
    * Limited data integrity, security, and scalability.
    * Poor performance when handling large or complex datasets.
    * Limited ability to manage relationships between data tables.


<span style="color: violet;">3. What are the advantages and disadvantages of using a database for health data management?</span>
* Advantages:
   * Efficient for managing large and complex datasets.
   * Supports data integrity, multi-user access, and security features.
   * Enables advanced data analysis and querying through relationships between tables.

* Disadvantages: 
   * Requires specialized knowledge (e.g., SQL, database design).
   * Higher setup and maintenance effort.
   * More complex to use compared to spreadsheets.


<span style="color: violet;">4. What are some examples of tasks that can be performed using a spreadsheet?</span>
* Data entry and formatting.
* Sorting, filtering, and summarizing data (e.g., calculating averages).
* Creating charts and graphs for data visualization.
* Performing calculations using formulas.



<span style="color: violet;">5. What are some examples of tasks that can be performed using a database?</span>
* Storing and managing large datasets with multiple relationships.
* Running complex queries (e.g., retrieving records based on multiple conditions).
* Ensuring data consistency and enforcing rules (e.g., primary keys, constraints).
* Handling multi-user access with different permissions.


<span style="color: violet;">6. What are some examples of tasks that can be performed using both a spreadsheet and a database?</span>
* Storing structured data (e.g., patient records, financial data).
* Performing basic data analysis (e.g., summarizing values or calculating totals).
* Filtering and sorting data based on criteria.
* Importing/exporting data to and from other formats (e.g., CSV).


<span style="color: violet;">7. What are some examples of tasks that can be performed using neither a spreadsheet nor a database? </span>
* Unstructured data analysis (e.g., analyzing free-text notes or images).
* Real-time data streaming (e.g., live monitoring of IoT devices).
* Complex statistical modeling that requires specialized software (e.g., machine learning).


<span style="color: violet;">8. What are some examples of tasks that can be performed using a spreadsheet but not a database?</span>
* Manual data entry and immediate ad-hoc analysis.
* Creating customized, interactive visualizations with minimal effort.
* Simple "what-if" analysis using tools like Excel Solver.


<span style="color: violet;">9. What are some examples of tasks that can be performed using a database but not a spreadsheet?</span>
* Managing large datasets with multiple interconnected tables.
* Enforcing strict data integrity through relationships, constraints, and triggers.
* Allowing simultaneous access and updates by multiple users with role-based permissions.


<span style="color: violet;">10. What are some examples of tasks that can be performed using both a spreadsheet and a database?</span>
* Data storage, retrieval, and basic reporting.
* Data cleaning and preparation (e.g., handling missing values).
* Exporting data to common formats like CSV for further processing.
* Simple data analysis (e.g., summarizing data by categories)
