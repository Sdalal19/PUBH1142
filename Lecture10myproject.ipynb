### <span style="color: Lightblue;">Data Visualization </span>

### Relational Plot

```python
# Use seaborn to create a scatter plot of Maternal Age versus Edinburg Postnatal Depression Scale (EPDS) score
seaborn.scatterplot(x='Maternal_Age', y='Edinburgh_Postnatal_Depression_Scale', data=df)
```

```python
# Recreate the plot with titles 
seaborn.scatterplot(
    data=df,
    x='Maternal_Age',
    y='Edinburgh_Postnatal_Depression_Scale'
);
matplotlib.pyplot.title('Age vs. EDPS score');
matplotlib.pyplot.xlabel('Age (Years)');
matplotlib.pyplot.ylabel('EPDS Score (1-30)');
matplotlib.pyplot.show()
```

```python
# add a hue to the graph above showing NICU stay yes and no to the scatter plot
seaborn.scatterplot(
    data=df,
    x='Maternal_Age',
    y='Edinburgh_Postnatal_Depression_Scale',
    hue='NICU_Stay'
);
matplotlib.pyplot.title('Age vs. EDPS score');
matplotlib.pyplot.xlabel('Age (Years)');
matplotlib.pyplot.ylabel('EPDS Score (1-30)');
matplotlib.pyplot.show()
```

```python
# add a style to the graph above showing NICU stay yes and no to the scatter plot
seaborn.scatterplot(
    data=df,
    x='Maternal_Age',
    y='Edinburgh_Postnatal_Depression_Scale',
    hue='NICU_Stay',
    style='NICU_Stay'
);
matplotlib.pyplot.title('Age vs. EDPS score');
matplotlib.pyplot.xlabel('Age (Years)');
matplotlib.pyplot.ylabel('EPDS Score (1-30)');
matplotlib.pyplot.show()
```

### Distribution - histogram

```python
seaborn.displot(
    df,
    x='Maternal_Age'
).set(
    title='Age distribution',
    xlabel='Age [years]',
    ylabel='Frequency'
);
```

```python
seaborn.displot(
    df,
    x='Maternal_Age',
    bins=range(10, 110, 5)
).set(
    title='Age distribution',
    xlabel='Age [years]',
    ylabel='Frequency'
);
```

```python
seaborn.displot(
    df,
    x='Maternal_Age',
    bins=range(10, 50, 5)
).set(
    title='Age distribution',
    xlabel='Age [years]',
    ylabel='Frequency'
);
```

```python
seaborn.displot(
    df,
    x='Maternal_Age',
    bins=range(10, 50, 3)
).set(
    title='Age distribution',
    xlabel='Age [years]',
    ylabel='Frequency'
);
```

```python
# distiguish the variables using the hue 
seaborn.displot(
    df,
    x='Maternal_Age',
    bins=range(10, 50, 3),
    hue='NICU_Stay',
).set(
    title='Age distribution',
    xlabel='Age [years]',
    ylabel='Frequency'
)

matplotlib.pyplot.legend(
    title='NICU_stay'
);
```

```python
# do the stacking 
seaborn.displot(
    df,
    x='Maternal_Age',
    bins=range(10, 50, 3),
    hue='NICU_Stay',
    multiple='stack'
).set(
    title='Age distribution',
    xlabel='Age [years]',
    ylabel='Frequency'
)

matplotlib.pyplot.legend(
    title='NICU_stay'
);
```

```python
# seperate the two plots
seaborn.displot(
    df,
    x='Maternal_Age',
    bins=range(10, 50, 3),
    col='NICU_Stay'
);
```

```python
# Fix the above plot by making the relative frequency version of the figure above is visualized by propability instead of frequency
seaborn.displot(
    df,
    x='Maternal_Age',
    bins=range(10, 50, 3),
    hue='NICU_Stay',
    col='NICU_Stay',
    stat='density'
).set(
    title='Age distribution',
    xlabel='Age [years]',
    ylabel='Frequency'
)
```

```python
seaborn.displot(
    df,
    x='Maternal_Age',
    y='Edinburgh_Postnatal_Depression_Scale',
    cbar=True
).set(
    title='Maternal Age vs. EDPS score',
    xlabel='Age [years]',
    ylabel='EPDS score'
);
```

```python
seaborn.displot(
    df,
    x='Maternal_Age',
    y='Edinburgh_Postnatal_Depression_Scale',
    kind='kde',
    rug=True
).set(
    title='Maternal Age vs. EDPS score',
    xlabel='Age [years]',
    ylabel='EPDS score'
);
```

```python
seaborn.jointplot(
    df,
    x='Maternal_Age',
    y='Edinburgh_Postnatal_Depression_Scale',
);
```

```python
g = seaborn.JointGrid(
    data=df,
    x='Maternal_Age',
    y='Edinburgh_Postnatal_Depression_Scale'
)
g.plot_joint(seaborn.histplot)
g.plot_marginals(seaborn.boxplot);
```

### Categorical Plots 

```python
seaborn.catplot(
    data=df,
    x='NICU_Stay',
    y='Maternal_Age',
    kind='box'
).set(
    title='Age and NICU stay',
    xlabel='NICU',
    ylabel='Age [years]'
);
```

```python
seaborn.catplot(
    data=df,
    x='NICU_Stay',
    y='Maternal_Age',
    hue='Delivery_Mode',
    kind='box'
).set(
    title='Age and NICU stay by Delivery Mode',
    xlabel='NICU stay',
    ylabel='Age [years]'
);
```

```python
# use the boxen to plot the same variables and plot above
seaborn.catplot(
    data=df,
    x='NICU_Stay',
    y='Maternal_Age',
    hue='Delivery_Mode',
    kind='boxen'
).set(
    title='Age and NICU stay by Delivery Mode',
    xlabel='NICU stay',
    ylabel='Age [years]'
);
```

```python
# create a point plot  for the difference in the Maternal_Age variable between those with and without NICU_stay, for each of the Vaginally delivery and C section Delivery, is demonstrated. The two categorical variables being NICU_Stay and Delivery_Mode
seaborn.catplot(
    data=df,
    x='NICU_Stay',
    y='Maternal_Age',
    hue='Delivery_Mode',
    kind='point'
).set(
    title='Age and NICU stay by Delivery Mode',
    xlabel='NICU stay',
    ylabel='Age [years]'
);
    
```
