### Distrubution - histogram

```python
seaborn.displot(
    df,
    x='age'
).set(
    title='Age distribution',
    xlabel='Age [years]',
    ylabel='Frequency'
);
```

```python
seaborn.displot(
    df,
    x='age',
    bins=range(40, 110, 10)
).set(
    title='Age distribution',
    xlabel='Age [years]',
    ylabel='Frequency'
);
```

Task: Generate a histogram to visualize the distribution of the platelets variable. Calculate the minimum and the maximum value for the variable and use the range function to create a intervals of 50000. Note that the ticks value on the horizontal axis seems overcrowded. Try to use the argument y instead of x. This will create a horizontal histogram, with more space for the tick values.

```python
# create a horizontal histogram
df.platelets.min()

df.platelets.max()

seaborn.displot(
    df,
    y='platelets',
    bins=range(0, 850000,50000)
).set(
    title='Platelet distribution',
    ylabel='Platelets [platelets/mL]',
    xlabel='Frequency'
);
```

```python
# distiguish the variables using the hue 
seaborn.displot(
    df,
    x='age',
    bins=range(40, 110, 10),
    hue='anaemia'
).set(
    title='Age distribution',
    xlabel='Age [years]',
    ylabel='Frequency'
)

matplotlib.pyplot.legend(
    title='Anemia'
);
```

```python
# Overlaying histograms can be difficult to visualize.
# Use the multiple parameter

seaborn.displot(
    df,
    x='age',
    bins=range(40, 110, 10),
    hue='anaemia',
    multiple='stack'
).set(
    title='Age distribution',
    xlabel='Age [years]',
    ylabel='Frequency'
);
```

```python
# seperate the two plots
seaborn.displot(
    df,
    x='age',
    bins=range(40, 110, 10),
    col='anaemia'
);
```

```python
# make relative frequency version of the figure above is visualized or colled a propability instead of frequency

seaborn.displot(
    df,
    x='age',
    bins=range(40, 110, 10),
    col='anaemia',
    stat='probability'
);
```

```python
# make a heat map of age and the platelets

seaborn.displot(
    df,
    x='age',
    y='platelets',
    cbar=True
).set(
    title='Age and Platelets',
    xlabel='Age [years]',
    ylabel='Platelets [platelets/mL]'
);
```

```python
# create a density model
seaborn.displot(
    df,
    x='age',
    y='platelets',
    kind='kde',
    rug=True
).set(
    title='Age and Platelets',
    xlabel='Age [years]',
    ylabel='Platelets [platelets/mL]'
);
```

```python
# combine them

seaborn.jointplot(
    df,
    x='age',
    y='platelets'
);
```

```python
# add box-and-whisker plots are added to the margin of a heat map.

g = seaborn.JointGrid(
    data=df,
    x='age',
    y='platelets'
)
g.plot_joint(seaborn.histplot)
g.plot_marginals(seaborn.boxplot);
```

### Categorical Plots 

```python
# a box-and-whisker plot of the age variable is created, for those with and without diabetes (unique elements in the diabetes variable)

seaborn.catplot(
    data=df,
    x='diabetes',
    y='age',
    kind='box'
).set(
    title='Age and Diabetes',
    xlabel='Diabetes',
    ylabel='Age [years]'
);
```

```python
# the box-and-whisker plot of the age variable is created, for those with and without diabetes (unique elements in the diabetes variable)
seaborn.catplot(
    data=df,
    x='diabetes',
    y='age',
    hue='death',
    kind='box'
).set(
    title='Age and Diabetes by Survival Group',
    xlabel='Diabetes',
    ylabel='Age [years]'
);
```

```python
# use the boxen to plot the same variables and plot above
seaborn.catplot(
    data=df,
    x='diabetes',
    y='age',
    kind='boxen'
).set(
    title='Age and Diabetes',
    xlabel='Diabetes',
    ylabel='Age [years]'
);
```

```python
seaborn.catplot(
    data=df,
    x='diabetes',
    y='age',
    hue='death',
    kind='boxen'
).set(
    title='Age and Diabetes by Survival Group',
    xlabel='Diabetes',
    ylabel='Age [years]'
);
```

```python
# turn the above plot into a violin plot
seaborn.catplot(
    data=df,
    x='diabetes',
    y='age',
    hue='death',
    kind='violin'
).set(
    title='Age and Diabetes by Survival Group',
    xlabel='Diabetes',
    ylabel='Age [years]'
);
```

```python
 #reduce the number of shapes, the violin plots can be split by the unique values 
seaborn.catplot(
    data=df,
    x='diabetes',
    y='age',
    hue='death',
    kind='violin',
    split=True
).set(
    title='Age and Diabetes by Survival Group',
    xlabel='Diabetes',
    ylabel='Age [years]'
);
```

```python
# create a swarm plot using the plot above
seaborn.catplot(
    data=df,
    x='diabetes',
    y='age',
    hue='death',
    kind='swarm'
).set(
    title='Age and Diabetes by Survival Group',
    xlabel='Diabetes',
    ylabel='Age [years]'
);
```

```python
# create a bar plot using the plot above
# have the frequency of those with and without diabetes is visualized
seaborn.catplot(
    data=df,
    x='diabetes',
    kind='count'
).set(
    title='Diabetes Frequency',
    xlabel='Diabetes',
    ylabel='Frequency'
);
```

```python
# color code the yes and the no
seaborn.catplot(
    data=df,
    x='diabetes',
    kind='count',
    palette='coolwarm'
).set(
    title='Diabetes Frequency',
    xlabel='Diabetes',
    ylabel='Frequency'
);
```

# Do a T-test

```python
# Do a students t test (using the scipy stats module) to determine if the age of those with and without diabetes is significantly different
from scipy import stats
stats.ttest_ind(
    df[df.diabetes == 'Yes'].age,
    df[df.diabetes == 'No'].age
)
```
