### <span style="color: lightpink;"> Comparing the relationship between Language and EPDS, Anxiety Score, Gestational Age, and Birth Weight </span>

#### <span style="color: violet;">Statistical Analysis (Independent T Test):</span>

```python
# Filter the data for English and French speakers
english_speakers = df[df['Language'] == 'English']['Edinburgh_Postnatal_Depression_Scale']
french_speakers = df[df['Language'] == 'French']['Edinburgh_Postnatal_Depression_Scale']

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(english_speakers, french_speakers, nan_policy='omit')

# Print the t-statistic and p-value
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

```python
#Do the same as above but for the Anxiety scores
english_speakers = df[df['Language'] == 'English']['PROMIS_Anxiety']
french_speakers = df[df['Language'] == 'French']['PROMIS_Anxiety']
# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(english_speakers, french_speakers, nan_policy='omit')
# Print the t-statistic and p-value
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

```python
#Do the same as above but for the Gestational Age at Birth
english_speakers = df[df['Language'] == 'English']['Gestational_Age_At_Birth']
french_speakers = df[df['Language'] == 'French']['Gestational_Age_At_Birth']
# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(english_speakers, french_speakers, nan_policy='omit')
# Print the t-statistic and p-value 
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

```python
#Do the same as above but for the Birth Weight
english_speakers = df[df['Language'] == 'English']['Birth_Weight']
french_speakers = df[df['Language'] == 'French']['Birth_Weight']
# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(english_speakers, french_speakers, nan_policy='omit')
# Print the t-statistic and p-value
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

## <span style="color: violet;">Conclusion:</span>

#### <span style="color: lightpink;">Although language was significantly associated with EPDS, anxiety scores, and birth weight in both T Tests and regression models, its predictive power was minimal, with very low RÂ² values in unadjusted analyses.
