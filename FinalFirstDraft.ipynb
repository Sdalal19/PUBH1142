##### <span style="color: teal;">The COVID-19 pandemic posed significant challenges to maternal health worldwide, amplifying stress, anxiety, and depression among pregnant and postpartum individuals. These mental health challenges were not experienced equallyâ€”language barriers often exacerbated health disparities. Non-English speaking individuals may have faced reduced access to information, limited communication with healthcare providers, and increased social isolation, all of which can intensify psychological distress.<span>

##### <span style="color: teal;">Language can serve as a proxy for broader cultural and structural inequalities, influencing not only maternal mental health but also birth outcomes such as gestational age, birth weight, and NICU admission rates. Despite its importance, language-based disparities in perinatal health have been underexplored, particularly in the context of large-scale public health crises like COVID-19. <span>

##### <span style="color: teal;">By examining differences in mental health scores and birth outcomes between English and non-English speaking participants, this study aims to shed light on the intersection between linguistic marginalization and maternal-infant wellbeing, with implications for equitable care and policy design.<span>
